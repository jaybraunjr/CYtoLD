```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
df = pd.read_csv('../arfgap1/output_all_residues.csv')

# Select residues 1-21
all_residues = df.columns[1:22]

# Define hydrophobic residues
hydrophobic_residues = ['ALA', 'VAL', 'LEU', 'ILE', 'MET', 'PHE', 'TRP', 'PRO']

# Exclude ARG and LYS residues
filtered_residues = [r for r in all_residues if r.split()[1] not in ['ARG', 'LYS']]

# Separate hydrophobic and other residues
other_residues = [r for r in filtered_residues if r.split()[1] not in hydrophobic_residues]
hydrophobic_filtered = [r for r in filtered_residues if r.split()[1] in hydrophobic_residues]

# Calculate the rolling average for each residue with a window size of 15
rolling_avg_df = df.rolling(window=15).mean()

# Plot the data
fig, ax = plt.subplots(figsize=(10, 6))

# Plot other residues first, in blue with a dotted line
for residue in other_residues:
    ax.plot(rolling_avg_df['time(ps)'], rolling_avg_df[residue], color='blue', marker='.',linestyle='None', markersize=4, label=residue)

# Plot hydrophobic residues after, in red with a dotted line
for residue in hydrophobic_filtered:
    ax.plot(rolling_avg_df['time(ps)'], rolling_avg_df[residue], color='red', marker='.',linestyle='None', markersize=4, label=residue)

# Customize plot labels and title
ax.set_ylabel('z-coordinate (nm)')
ax.set_title('Depth Analysis for residues 1-21')

ax.set_ylim([-10, 60])  # Set y-limit here
# Add legend
ax.legend()

plt.show()
```
