```python
df = pd.read_csv('rep3/rolling.csv')
x = df['time']
```

```python
colors = {19:'b',20:'b',21:'r',22:'r',23:'b',24:'r',25:'b',26:'r',28:'r',29:'r',30:'b',31:'b',32:'r',33:'b',34:'b',35:'b',36:'r',37:'b'}
line1, = plt.plot(x,df[f'resid_{19}'], marker='', linestyle='-', markersize=2, color=colors[19])
line2, = plt.plot(x,df[f'resid_{20}'], marker='', linestyle='-', markersize=2, color='r')
for resid in range(21, 37):
#     if resid == 22:
#         continue
    if resid == 27:
        continue
    plt.plot(x,df[f'resid_{resid}'], marker='.', linestyle='None', markersize=2, color=colors[resid])

plt.legend([line1,line2],['polar/charged', 'hydrophobic'], title='Residue Type',
           bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.,
           markerfirst=False,
           labelspacing=0.5,
           handletextpad=0.5,
           handlelength=2,
           handleheight=2,
           fancybox=True,
           shadow=True,
           facecolor='white',
           edgecolor='blue',
           framealpha=0.5)
plt.axhline(y=0.0, color='black', linestyle='-')
plt.ylabel('depth ($\AA$)')
plt.ylim(-10,20)
```
