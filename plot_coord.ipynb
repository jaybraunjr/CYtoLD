```python
df = pd.read_csv('arg_pl.csv')
df2 = pd.read_csv('arg_pl_err.csv')
df3 = pd.read_csv('lys_pl.csv')
df4 = pd.read_csv('lys_pl_err.csv')

x = ['HG','PO4','GL','C1','C2','C3','C4']
fig, ax = plt.subplots(figsize=(6, 3))

ax.scatter(x,df['0'],c='tab:blue',marker='s',linewidth=5)
plt.plot(x, df['0'], c='tab:blue')

ax.scatter(x,df3['0'],c='tab:orange',marker='s',linewidth=5)
plt.plot(x, df3['0'], c='tab:orange')


plt.title("ARG (229) and LYS (240) to PL coordination")
plt.ylabel("||s||")

ax.errorbar(x, df['0'], yerr=df2['0'], fmt='o', c='tab:blue', capsize=3, capthick=0.5, uplims=True)
ax.errorbar(x, df3['0'], yerr=df4['0'], fmt='o', c='tab:orange', capsize=3, capthick=0.5, uplims=True)
plt.xticks(rotation=45)
# plt.savefig('leu_coord.png',dpi=200)
plt.show()
```

```python
df1 = pd.read_csv('leu228_pl.csv')
df2 = pd.read_csv('leu228_pl_err.csv')

df3 = pd.read_csv('val232_pl.csv')
df4 = pd.read_csv('val232_pl_err.csv')

df5 = pd.read_csv('leu236_pl.csv')
df6 = pd.read_csv('leu236_pl_err.csv')

df7 = pd.read_csv('leu243_pl.csv')
df8 = pd.read_csv('leu243_pl_err.csv')



x_labels = ['HG','PO4','GL','C1','C2','C3','C4']
x = range(len(x_labels))
x_offset = [i - 0.2 for i in x]
x_offset2 = [i + 0.2 for i in x]
x_offset3 = [i + 0.4 for i in x]

fig, ax = plt.subplots(figsize=(6, 3))

ax.scatter(x, df1['0'], c='tab:blue', marker='s', linewidth=5, label='leu(228)')
plt.plot(x, df1['0'], c='tab:blue')

ax.scatter(x_offset, df3['0'], c='tab:orange', marker='s', linewidth=5,label = 'val(232)')
plt.plot(x_offset, df3['0'], c='tab:orange')

ax.scatter(x_offset2, df5['0'], c='tab:purple', marker='s', linewidth=5, label='leu(236)')
plt.plot(x_offset2, df5['0'], c='tab:purple')

ax.scatter(x_offset3, df7['0'], c='tab:green', marker='s', linewidth=5,label = 'leu(243)')
plt.plot(x_offset3, df7['0'], c='tab:green')

plt.title("conserved hydrophobic residues (TG-LD)")
plt.ylabel("coordination")

ax.errorbar(x, df1['0'], yerr=df2['0'], fmt='o', c='tab:blue', capsize=3, capthick=0.5, uplims=True)
ax.errorbar(x_offset, df3['0'], yerr=df4['0'], fmt='o', c='tab:orange', capsize=3, capthick=0.5, uplims=True)
ax.errorbar(x_offset2, df5['0'], yerr=df6['0'], fmt='o', c='tab:purple', capsize=3, capthick=0.5, uplims=True)
ax.errorbar(x_offset3, df7['0'], yerr=df8['0'], fmt='o', c='tab:green', capsize=3, capthick=0.5, uplims=True)



plt.xticks(x, x_labels, rotation=45)
plt.legend()
plt.savefig('coord.png',dpi=300)
plt.show()
```

```python
df1 = pd.read_csv('arg229_pl.csv')
df2 = pd.read_csv('arg229_pl_err.csv')

df3 = pd.read_csv('lys240_pl.csv')
df4 = pd.read_csv('lys240_pl_err.csv')




x_labels = ['HG','PO4','GL','C1','C2','C3','C4']
x = range(len(x_labels))
x_offset = [i - 0.2 for i in x]
x_offset2 = [i + 0.2 for i in x]
x_offset3 = [i + 0.4 for i in x]

fig, ax = plt.subplots(figsize=(6, 3))

ax.scatter(x, df1['0'], c='tab:blue', marker='s', linewidth=5, label='arg(229)')
plt.plot(x, df1['0'], c='tab:blue')

ax.scatter(x_offset, df3['0'], c='tab:orange', marker='s', linewidth=5,label = 'lys(240)')
plt.plot(x_offset, df3['0'], c='tab:orange')


plt.title("charged residues (TG-LD)")
plt.ylabel("coordination")

ax.errorbar(x, df1['0'], yerr=df2['0'], fmt='o', c='tab:blue', capsize=3, capthick=0.5, uplims=True)
ax.errorbar(x_offset, df3['0'], yerr=df4['0'], fmt='o', c='tab:orange', capsize=3, capthick=0.5, uplims=True)

plt.legend()
plt.xticks(x, x_labels, rotation=45)
plt.savefig('coord_charged.png',dpi=300)
plt.show()
```
